## User-Defined Helper Functions
This section is a collection of user-defined helper functions that are used all along this notebook. A better way to showcase these user-defined functions is to wrap them in a python files and import functions from the file. For this challenge, we show all the function in this notebook itself. 

```python
def get_address_from_lat_lng(lat, lng):
    """
        Returns a Country for a set of lat, lng (inputs) 
        - using GeoPy Nominatm python library
    """
    geolocator = Nominatim(user_agent="krikey_challenge")
    # Location
    location = geolocator.reverse([str(lat), str(lng)])

    # Country
    country = location.raw['address']['country']
    
    return country


def plot_dots_folium(point):
    """
        Returns set of circle markers on top of base folium map 
    """
    folium.CircleMarker(location=[point.lat,point.lng],
                       radius=5,
                       weight=2,
                       popup=[point.lat, point.lng, point.id],
                       fill_color='#000000').add_to(m)
```

## Exploratory Data Analysis

### Correcting datatypes and Feature engineering of user attributes

```python
# 1. convert "created_at" to datetime 
df_users['created_at'] = pd.to_datetime(df_users['created_at'], format='%Y-%m-%d %H:%M:%S')

# 2.1 "coordinates" into 2 attributes (latitude-lat and longitude-lng)
df_users[["lat", "lng"]] = df_users["coordinates"].str.split(",", expand=True)
df_users['lat'] = df_users['lat'].str.replace("(",'', regex=True)
df_users['lng'] = df_users['lng'].str.replace(")",'', regex=True)

# 2.2 convert 2 new attributes to floats
df_users['lat'] = df_users['lat'].apply(lambda x: np.float(x.strip().replace("'",'')))
df_users['lng'] = df_users['lng'].apply(lambda x: np.float(x.strip().replace("'",'')))

#df_users['country'] = df_users.apply(lambda x: get_address_from_lat_lng(x.lat, x.lng), axis=1)

df_users.head()
```

#### How many users are there by class?
Here, we calculate no. of user per class. That is, we get information such as how many users are creators, gamers, super creators, super gamers, or viewers. 

```python
# Create dataframe for no. of users per class
df_users_class = df_users.groupby('class')['id'].count().reset_index()
df_users_class.columns = ['user_class','user_count']
df_users_class
```

```python
# Plot no. of users per class
df_users_class.plot_bokeh(kind='bar', title='No. of users by class',
                    x = 'user_class',
                    xlabel='Class of the User',
                    ylabel='No. of Users',
                    legend='top_left', alpha=0.6);
```

#### How many users are there by country or geographic location?
To get no. of users per country, we first have to convert user registered lat, lng into country attribute. This is done by using geopy's Nominatm library. The function `get_address_from_lat_lng` takes lat, lng as input and converts them into a address and returns a country. We limit geographical analysis to country level for this challenge and plot top 10 countries by users.   

```python
# Create dataframe for no. of users per lat, lng
df_users_country = df_users.groupby(['lat','lng'])['id'].count().reset_index()
df_users_country['country'] = ''
df_users_country.head()
```

Above takes indicates that users registered from around **626** unique locations. Now, we loop over above `df_users_country` dataframe to determine the country for every unique lat, lng. Since there are ~600 locations, converting all those lat, lng to country might take few minutes. To avoid accidently running this code again, we comment them after getting all countries for lat, lng and saving them in a seperate csv file. 

```python
# Country from lat,lng
for i in range(df_users_country.shape[0]):
    lat = df_users_country.loc[i]['lat']
    lng = df_users_country.loc[i]['lng']
    country = get_address_from_lat_lng(lat, lng)
    #df_users_country.loc[i, 'country'] = country
    if (i % 25 == 0):
        print(i, lat, lng)#, country)
```

```python
# Save dataframe 
df_users_country.to_csv('lat_lng_country.csv')
```

```python
# Top 10 countries by users
df_users_country_all = df_users_country.groupby(['country'])['id'].count().reset_index()
df_users_country_all.columns = ['country','user_count']
df_users_country_top10 = df_users_country_all.sort_values(by='user_count', ascending=False).head(10)
print(df_users_country_all.shape[0])
df_users_country_top10
```

```python
# User Count by lat, lng -> for mapplot
df_users_count_lat_lng = df_users_country.groupby(['lat','lng'])['id'].count().reset_index()
df_users_count_lat_lng
```

```python
# Pie chart of Top10 countries by users
df_users_country_top10.plot_bokeh(kind='pie', x="country", y="user_count", 
                                  title="Top 10 countries by user base",
                                  colormap=["blue", "red", "yellow", "green", "purple", "orange", "grey",'pink','brown','lightblue']
                                 );
```

```python
# Plot No. of Users by country using Folium library 
m = folium.Map(tiles="StamenToner", min_zoom=2)

# No. of users and lat, lng
nusers = df_users_count_lat_lng['id'].astype(float)
lat = df_users_count_lat_lng['lat'].astype(float)
lng = df_users_count_lat_lng['lng'].astype(float)

# Add nusers data as Heatmap 
m.add_child(HeatMap(zip(lat,lng, nusers),radius=0))

# Add dot per user on top of Heatmap 
df_users_count_lat_lng.apply(plot_dots_folium, axis=1)
m.fit_bounds(m.get_bounds())
m
```

```python
# Plot Folium map 
m = folium.Map(zoom_start=11)

# Add additional Tiles
folium.TileLayer('Stamen Terrain').add_to(m)
folium.TileLayer('Stamen Toner').add_to(m)
folium.TileLayer('Stamen Water Color').add_to(m)
folium.TileLayer('cartodbpositron').add_to(m)

# Add marker cluster
marker_cluster = MarkerCluster(name='No. of users clustered icons',
                               overlay=True, control=False, icon_create_function=None)

for k in range(len(lat)):
    location = lat[k], lng[k]
    marker = folium.Marker(location=location)
    popup = 'lng:{}<br> lat:{}<br> user_count:{}'.format(location[1], location[0], nusers[k])
    folium.Popup(popup).add_to(marker)
    marker_cluster.add_child(marker)

marker_cluster.add_to(m)

folium.LayerControl().add_to(m);
m
```

### How many users have registered over date and time?
To answer this question, we can make use of the `created_at` attribute to calculate the no. of users over date and time. 
1. First, we compute no. of users registered every month during 2021.
2. Next, we find no. of users registered by day of week
3. Finally, we compute no. of users registered by hour of day.

#### Table of no. of users by month

```python
# No. of users registered per month in 2021
df_users_month = df_users.groupby(df_users['created_at'].dt.month)['id'].count().reset_index()
df_users_month.columns = ['Month', 'user_count']

# Convert month number to Month name
df_users_month['Month'] = df_users_month['Month'].apply(lambda x: calendar.month_abbr[x])
df_users_month
```

#### Plot of no. of users by month

```python
# Line of no. of users by month in 2021
df_users_month.plot_bokeh(kind='line', figsize=(800, 450), title="Monthly user registrations in 2021",
              xlabel="Months in 2021", ylabel="No. of users",
              x='Month', y='user_count',
              ylim=(0, 150),
              plot_data_points=True, plot_data_points_size=10, marker="o");
```

#### Table of no. of users by Day of the week

```python
# No. of users registered by day of week
df_users_dow = df_users.groupby(df_users['created_at'].dt.dayofweek)['id'].count().reset_index()
df_users_dow.columns = ['day_of_week', 'user_count']

# Convert DOW number to DOW name
df_users_dow['day_of_week'] = df_users_dow['day_of_week'].apply(lambda x: calendar.day_abbr[x])
df_users_dow
```

#### Plot of no. of users by Day of the week

```python
# Line of no. of users by Day of week
df_users_dow.plot_bokeh(kind='line', figsize=(800, 450), title="Days of the week user registrations in 2021",
              xlabel="Days of the Week", ylabel="No. of users",
              x='day_of_week', y='user_count',
              ylim=(0, 200),
              plot_data_points=True, plot_data_points_size=10, marker="o");
```

#### Table of no. of users by Hour of the day

```python
# No. of users registered by hour of the day
df_users_hr = df_users.groupby(df_users['created_at'].dt.hour)['id'].count().reset_index()
df_users_hr.columns = ['hour_of_day', 'user_count']

# Create hour category
df_users_hr['hr_category'] = pd.cut(df_users_hr['hour_of_day']+1, np.arange(0, 27, 3))

# create users_hr_category
df_users_hr_cat = df_users_hr.groupby('hr_category')['user_count'].sum().reset_index()
df_users_hr_cat['hr_category'] = df_users_hr_cat['hr_category'].apply(lambda x: str(x))
df_users_hr_cat['user_count_pct'] = df_users_hr_cat['user_count'] / df_users_hr_cat['user_count'].sum() * 100
df_users_hr_cat
```

#### Plot of no. of users by Hour of the day

```python
# Plot no. of users by hour of the day 
df_users_hr_cat.plot_bokeh(kind='pie', x="hr_category", y='user_count_pct', title="Percent of users by hour of the day",
                       colormap=["blue", "red", "yellow", "green", "purple", "orange", "grey", 'pink'],
                       line_color="grey");
```
